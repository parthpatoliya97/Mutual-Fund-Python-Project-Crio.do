```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('nifty50_closing_prices.csv')

df['Date'] = pd.to_datetime(df['Date'])

# Set Date as index
df.set_index('Date', inplace=True)

df.head()
```

```python
df.isna().sum()
```

```python
daily_returns = df.pct_change().dropna()
daily_returns.head()
```

```python
roi = (df.iloc[-1] / df.iloc[0] - 1) * 100
roi = roi.sort_values(ascending=False)

roi.head(10)
```

```python
volatility = daily_returns.std() * np.sqrt(252) * 100
volatility = volatility.sort_values()

volatility.head(10)
```

```python
metrics = pd.DataFrame({
    'ROI (%)': roi,
    'Volatility (%)': volatility
})

metrics.dropna(inplace=True)
metrics.head()
```

```python
roi_threshold = metrics['ROI (%)'].quantile(0.60)
vol_threshold = metrics['Volatility (%)'].quantile(0.40)
```

```python
selected_stocks = metrics[
    (metrics['ROI (%)'] >= roi_threshold) &
    (metrics['Volatility (%)'] <= vol_threshold)
]

selected_stocks
```

```python
inv_vol = 1 / selected_stocks['Volatility (%)']
weights = inv_vol / inv_vol.sum()

selected_stocks['Weight'] = weights
selected_stocks
```

```python
portfolio_roi = np.sum(selected_stocks['ROI (%)'] * selected_stocks['Weight'])
portfolio_risk = np.sqrt(
    np.sum((selected_stocks['Weight'] ** 2) * (selected_stocks['Volatility (%)'] ** 2))
)

portfolio_roi, portfolio_risk
```

```python
high_growth_stocks = metrics.sort_values('ROI (%)', ascending=False).head(5)

high_growth_roi = high_growth_stocks['ROI (%)'].mean()
high_growth_risk = high_growth_stocks['Volatility (%)'].mean()

comparison = pd.DataFrame({
    'Category': ['Mutual Fund', 'High Growth Stocks'],
    'Average ROI (%)': [portfolio_roi, high_growth_roi],
    'Average Risk (%)': [portfolio_risk, high_growth_risk]
})

comparison
```

```python
monthly_investment = 5000
annual_increment = 0.10
expected_annual_return = portfolio_roi / 100
monthly_return = expected_annual_return / 12
```

```python
def sip_future_value(monthly_amt, years, annual_step_up, monthly_rate):
    total_months = years * 12
    value = 0
    current_monthly = monthly_amt
    
    for month in range(1, total_months + 1):
        if month % 12 == 1 and month > 1:
            current_monthly *= (1 + annual_step_up)
        
        value = (value + current_monthly) * (1 + monthly_rate)
    
    return round(value, 2)
```

```python
investment_periods = [1, 3, 5, 10]

results = {
    years: sip_future_value(
        monthly_investment,
        years,
        annual_increment,
        monthly_return
    )
    for years in investment_periods
}

results
```

```python
plt.figure(figsize=(8,5))
plt.plot(results.keys(), results.values(), marker='o')
plt.title("Mutual Fund SIP Growth Over Time")
plt.xlabel("Years")
plt.ylabel("Investment Value (₹)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
from datetime import datetime
```

```python
# Load data
df = pd.read_csv('nifty50_closing_prices.csv', parse_dates=['Date'], index_col='Date')

# Sort values
df = df.sort_index()

# Forward fill missing values
df = df.fillna(method='ffill')

# Calculate Daily Returns
returns = df.pct_change().dropna()

# Convert to Monthly Returns
monthly_returns = returns.resample('M').mean()
```

```python
# 1️⃣ Total Return
total_return = (df.iloc[-1] / df.iloc[0]) - 1
```

```python
# 2️⃣ Annualized Return
annualized_return = (1 + returns.mean()) ** 252 - 1
```

```python
# 3️⃣ Volatility (Risk)
annualized_volatility = returns.std() * np.sqrt(252)
```

```python
# 4️⃣ Sharpe Ratio
risk_free_rate = 0.06  # Assume 6% for India

sharpe_ratio = (annualized_return - risk_free_rate) / annualized_volatility
```

```python
# 5️⃣ Sortino Ratio
downside = returns[returns < 0]
downside_std = downside.std() * np.sqrt(252)

sortino_ratio = (annualized_return - risk_free_rate) / downside_std
```

```python
selected_stocks = returns.columns[:3]   # pick first 3 stocks
returns_3 = returns[selected_stocks]

weights = np.array([0.4, 0.3, 0.3])

# Annualized return
annualized_return_3 = (1 + returns_3.mean())**252 - 1

portfolio_return = np.sum(weights * annualized_return_3)

portfolio_volatility = np.sqrt(
    np.dot(weights.T, np.dot(returns_3.cov() * 252, weights))
)

print("Portfolio Return:", portfolio_return)
print("Portfolio Volatility:", portfolio_volatility)
```

```python
num_assets = len(returns.columns)

# Equal weights for all stocks
weights = np.array([1/num_assets] * num_assets)

annualized_return = (1 + returns.mean())**252 - 1

portfolio_return = np.sum(weights * annualized_return)

portfolio_volatility = np.sqrt(
    np.dot(weights.T, np.dot(returns.cov() * 252, weights))
)

print("Portfolio Return:", portfolio_return)
print("Portfolio Volatility:", portfolio_volatility)
```

```python
# STEP 5 — Monte Carlo Simulation
def monte_carlo_simulation(weights, returns, num_simulations=1000, num_days=252):
    
    portfolio_mean = np.sum(weights * returns.mean())
    portfolio_std = np.sqrt(np.dot(weights.T, np.dot(returns.cov(), weights)))

    simulation_results = np.zeros((num_days, num_simulations))

    for i in range(num_simulations):
        daily_returns = np.random.normal(portfolio_mean, portfolio_std, num_days)
        simulation_results[:, i] = np.cumprod(1 + daily_returns)

    return simulation_results
```

```python
sim_results = monte_carlo_simulation(weights, returns)

plt.figure(figsize=(10,6))
plt.plot(sim_results)
plt.title("Monte Carlo Simulation of Portfolio")
plt.show()
```

```python
# STEP 6 — Efficient Frontier (Mean-Variance Optimization)
def portfolio_performance(weights, mean_returns, cov_matrix):
    returns = np.sum(mean_returns * weights) * 252
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix * 252, weights)))
    return returns, std
```

```python
def minimize_volatility(weights):
    return portfolio_performance(weights, returns.mean(), returns.cov())[1]

constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
bounds = tuple((0,1) for _ in range(len(returns.columns)))

initial_weights = np.array([1/len(returns.columns)] * len(returns.columns))

optimal = minimize(minimize_volatility, initial_weights,
                   method='SLSQP', bounds=bounds, constraints=constraints)

optimal_weights = optimal.x
```

```python
num_portfolios = 5000
results = np.zeros((3, num_portfolios))

for i in range(num_portfolios):
    weights = np.random.random(len(returns.columns))
    weights /= np.sum(weights)
    
    port_return, port_std = portfolio_performance(weights, returns.mean(), returns.cov())
    
    results[0,i] = port_return
    results[1,i] = port_std
    results[2,i] = (port_return - risk_free_rate) / port_std

plt.figure(figsize=(10,6))
plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='viridis')
plt.xlabel("Volatility")
plt.ylabel("Return")
plt.colorbar(label='Sharpe Ratio')
plt.title("Efficient Frontier")
plt.show()
```

```python
# STEP 8 — Portfolio Composition Pie Chart
plt.figure(figsize=(6,6))
plt.barh(optimal_weights)
plt.title("Optimized Portfolio Allocation")
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Parameters
initial_monthly_investment = 5000  # Starting monthly investment in INR
investment_years = [1, 3, 5, 10]   # Investment periods
n = 12  # Monthly compounding
annual_increase_rate = 0.10  # 10% increase in monthly investment every year

# Replace this with the actual ROI array or value
expected_roi_mutual_fund = np.array([12.5, 13.2, 14.0])  # Example ROI values
avg_roi = expected_roi_mutual_fund.mean() / 100  # Convert to decimal

# Function to calculate future value with increasing annual investments
def future_value_with_growth(initial_P, r, n, years, growth_rate):
    total_fv = 0
    for year in range(years):
        # Calculate monthly investment for this year
        P = initial_P * ((1 + growth_rate) ** year)
        # Time left for this annual investment to compound
        t = years - year
        # Future value of this year's monthly SIPs
        fv = P * (((1 + r / n) ** (n * t) - 1) / (r / n)) * (1 + r / n)
        total_fv += fv
    return total_fv

# Calculate future values for each investment period
future_values = [future_value_with_growth(initial_monthly_investment, avg_roi, n, t, annual_increase_rate)
                 for t in investment_years]

# Plotting
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[f"{year} year{'s' if year > 1 else ''}" for year in investment_years],
    y=future_values,
    mode='lines+markers',
    line=dict(color='blue'),
    marker=dict(size=8),
    name='Future Value'
))

fig.update_layout(
    title="Expected Value of Monthly SIP with 10% Annual Increase (₹5000 Starting)",
    xaxis_title="Investment Period",
    yaxis_title="Future Value (INR)",
    xaxis=dict(showgrid=True, gridcolor='lightgrey'),
    yaxis=dict(showgrid=True, gridcolor='lightgrey'),
    template="plotly_white",
    hovermode='x unified'
)

fig.show()
```
